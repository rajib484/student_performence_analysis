```python
# comparing study level and life style 
import pandas as pd

# Remove any column space issues
my_file.columns = my_file.columns.str.strip()

# Find the index of the max and min GPA
max_index = my_file["GPA"].idxmax()
min_index = my_file["GPA"].idxmin()

# Get full student profiles
max_gpa = my_file.loc[max_index]
min_gpa = my_file.loc[min_index]

# Combine into a comparison DataFrame
compare = pd.DataFrame({
    "Category": [
        "Study Hours per Day",
        "Sleep Hours per Day",
        "Extracurricular Hours per Day",
        "Social Hours per Day",
        "Physical Activity Hours per Day",
        "Stress Level",
        "GPA"
    ],
    "Highest GPA Student": [
        max_gpa["Study_Hours_Per_Day"],
        max_gpa["Sleep_Hours_Per_Day"],
        max_gpa["Extracurricular_Hours_Per_Day"],
        max_gpa["Social_Hours_Per_Day"],
        max_gpa["Physical_Activity_Hours_Per_Day"],
        max_gpa["Stress_Level"],
        max_gpa["GPA"]
    ],
    "Lowest GPA Student": [
        min_gpa["Study_Hours_Per_Day"],
        min_gpa["Sleep_Hours_Per_Day"],
        min_gpa["Extracurricular_Hours_Per_Day"],
        min_gpa["Social_Hours_Per_Day"],
        min_gpa["Physical_Activity_Hours_Per_Day"],
        min_gpa["Stress_Level"],
        min_gpa["GPA"]
    ]
})

print(compare)
print(compare.dtypes)

compare["Highest GPA Student"] = pd.to_numeric(compare["Highest GPA Student"], errors='coerce')
compare["Lowest GPA Student"] = pd.to_numeric(compare["Lowest GPA Student"], errors='coerce')
print(compare.dtypes)
import matplotlib.pyplot as plt

compare.set_index("Category")[["Highest GPA Student", "Lowest GPA Student"]].plot(kind="bar")
plt.title("Comparison: Study Habits and Lifestyle (High vs Low GPA)")
plt.ylabel("Hours / Level")
plt.xticks(rotation=45)
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Clean column names (important if you have spaces)
my_file.columns = my_file.columns.str.strip()

# Select variables
X = my_file[["Study_Hours_Per_Day"]]   # independent variable
y = my_file["GPA"]                      # dependent variable

# Build and train the model
model = LinearRegression()
model.fit(X, y)

# Predict values
y_pred = model.predict(X)

# Plot actual data vs predicted line
plt.scatter(X, y, color='blue', label='Actual GPA')
plt.plot(X, y_pred, color='red', linewidth=2, label='Regression Line')

plt.title("Study Hours vs GPA (Linear Regression)")
plt.xlabel("Study Hours Per Day")
plt.ylabel("GPA")
plt.legend()
plt.show()

# Model information
print("Coefficient (Slope):", model.coef_[0])
print("Intercept:", model.intercept_)
```
